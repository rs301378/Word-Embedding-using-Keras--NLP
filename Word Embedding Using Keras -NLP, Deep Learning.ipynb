**Word Embedding Technniques using Enbedding layer in Keras**

```python
# Libraries Used: Keras, TensorFlow
```

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
#sentences
sent= ['the glass of milk',
      'the glass of juice',
      'the cup of tea',
      'i am a good boy',
      'I am a good developer',
      'understand the meaning of words',
      'your videos are good']
sent
```

```python
voc_size=10000  #size of the dictionary
```

**One Hot Representation**

```python
onehot_repr = [one_hot(words, voc_size) for words in sent]
print(onehot_repr)  #index from the dictionary
```

**Word Embedding Representation**

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
sent_length = 8
embedded_docs = pad_sequences(onehot_repr, padding = 'pre', maxlen = sent_length)
# 'pre' it will add zeros in front of sentences to make it equal
```

```python
print(embedded_docs)
```

```python
dim = 10  #features
```

```python
model = Sequential()
model.add(Embedding(voc_size, 10, input_length = sent_length))
model.compile('adam', 'mse')
```

```python
model.summary()
```

```python
#in order to see how my senteces will converted into vectors
print(model.predict(embedded_docs))
```
